```python
import pandas as pd


# Load the dataset
file_path = "/content/drive/MyDrive/BDM Capstone/Processed Raw data/Quantised Business Process and Operation Index - Sheet1.csv"
df = pd.read_csv(file_path)
```

```python


import matplotlib.pyplot as plt
import seaborn as sns

# Pairplot to visualize relationships between variables
sns.pairplot(df[['Relative Production Efficiency(1-10)', 'Relative Turnover(1-10)', 'Return Rate(1-5)', 'Customer Satisfaction(1-5)', 'Quality Assurance Process(1-5)', 'Relative Invetory Level(1-10)', 'Employee Satisfaction(1-5)', 'Supply Chain Effectiveness(1-5)', 'Strategy Effectiveness(1-5)', 'Business data Maintainance(1-10)', 'Financial Performance(1-10)', 'Innovation/ Technology Improvement(1-5)', 'Descripencies and Hurdles(1-10)', 'Marketing Efficiency(1-5)', 'Sustainability Score']], diag_kind='kde')
plt.show()
```

```python
# Calculate the correlation matrix
correlation_matrix = df.corr()

# Heatmap to visualize correlation matrix
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

```python
month_mapping = {
    'January': 1,
    'February': 2,
    'March': 3,
    'April': 4,
    'May': 5,
    'June': 6,
    'July': 7,
    'August': 8,
    'September': 9,
    'October': 10,
    'November': 11,
    'December': 12
}

# Convert 'month' column to numeric
df['month'] = df['month'].map(month_mapping)
# Convert 'month' and 'year' columns to a datetime format
df['date'] = pd.to_datetime(df[['month', 'year']].assign(DAY=1))


# Set the date as the index
df.set_index('date', inplace=True)

# Drop 'month' and 'year' columns
df.drop(['month', 'year'], axis=1, inplace=True)


# Plot sustainability score over time
plt.figure(figsize=(12, 6))
plt.plot(df['Sustainability Score'], marker='o', linestyle='-')
plt.title('Sustainability Score Over Time')
plt.xlabel('Date')
plt.ylabel('Sustainability Score')
plt.grid(True)
plt.show()
```
