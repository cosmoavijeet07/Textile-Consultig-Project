```python
# Create a column 'month_year' by combining 'year' and 'month_number'
sales_messo['month_year'] =  sales_messo['month_number'].astype(str) + '-' +sales_messo['financial_year'].astype(str)
```

```python
import calendar

def replace_month_numbers_with_names(dataframe, month_column_name):
    # Define a dictionary to map month numbers to month names
    month_name_map = {i: calendar.month_name[i] for i in range(1, 13)}

    # Extract month and year from the format 'M-YYYY' and replace with month name
    dataframe[month_column_name] = dataframe[month_column_name].apply(
        lambda x: month_name_map[int(x.split('-')[0])] + '-' + x.split('-')[1]
    )
```

```python
# Call the function to replace month numbers with month names
replace_month_numbers_with_names(sales_messo, 'month_year')
sales_messo.head()
```

```python
# Check for duplicates in 'Column1'
duplicates = sales_messo['sub_order_num'].duplicated(keep=False)

# Display the rows with duplicates
duplicated_rows = sales_messo[duplicates]
duplicated_rows[:7]
```

#### Month-year sales pivot

```python
# Group the data by 'month' and 'order_status' and sum the 'quantity'
monthly = sales_messo.groupby(['month_year', 'order_status'])['quantity'].sum().reset_index()

# Pivot the table to get 'order_status' as columns
pivot_table1 = monthly.pivot(index='month_year', columns='order_status', values='quantity').fillna(0)

# Add a new column 'Total' representing the total transaction quantity for each month
pivot_table1['Total'] = pivot_table1.sum(axis=1)

# If there are NaN values in the pivot_table, you can replace them with 0
pivot_table1 = pivot_table1.fillna(0)
```

```python
# Define a custom sorting order based on the month names
month_order = [
    'January-2022', 'February-2022', 'March-2022', 'April-2022', 'May-2022', 'June-2022',
    'July-2022', 'August-2022', 'September-2022', 'October-2022', 'November-2022', 'December-2022', 'January-2023', 'February-2023', 'March-2023', 'April-2023', 'May-2023', 'June-2023',
    'July-2023', 'August-2023', 'September-2023'
]
# Create a categorical data type with the custom sorting order
pivot_table1.index = pd.Categorical(pivot_table1.index, categories=month_order, ordered=True)

# Sort the pivot table based on the custom sorting order
Monthly_sales = pivot_table1.sort_index()
```

### Reports and Graphs

```python
Monthly_sales
```

```python
# Plot the dynamic line chart
for order_status in Monthly_sales.columns:
    plt.plot(Monthly_sales.index, Monthly_sales[order_status], label=order_status)

plt.xlabel('Month-year')
plt.ylabel('Quantity')
plt.title('Monthly Sales Quantity by Order Status')
plt.legend()
plt.xticks(rotation=45)  # Rotate x-axis labels for readability
plt.grid(True)

# Show the dynamic line chart
plt.show()
```

```python
# Create an interactive line chart using Plotly
fig = px.line(monthly, x='month_year', y='quantity', color='order_status',
              labels={'month': 'Month', 'quantity': 'Quantity'},
              title='Monthly Sales Quantity by Order Status')

# Show the interactive chart
fig.show()
```

```python
# Group by state and order status, summing up the quantity
import seaborn as sns
grouped_data = sales_messo.groupby(['end_customer_state_new', 'order_status']).sum().reset_index()

# Plotting using seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x='end_customer_state_new', y='quantity', hue='order_status', data=grouped_data)
plt.title('Quantity of Sells by State and Order Status')
plt.xlabel('State')
plt.ylabel('Quantity')
plt.show()
```

```python
import plotly.express as px
# Create an interactive bar chart using Plotly Express
fig = px.bar(grouped_data, y='end_customer_state_new', x='quantity', color='order_status',
             labels={'quantity': 'Quantity', 'end_customer_state_new': 'State'},
             title='Quantity of Sells by State and Order Status',
             barmode='group')

# Show the plot
fig.show()
```

```python
import geopandas as gpd
import requests
# # Load GeoJSON file for Indian states
# geojson_path = 'path/to/india_states.geojson'
# india_states = gpd.read_file(geojson_path) # Built-in GeoJSON for Indian states
# Load built-in GeoJSON for Indian states using Plotly Express
# india_states = px.data.india()

# Load GeoJSON file for Indian states from datameet GitHub repository
geojson_url = 'https://raw.githubusercontent.com/datameet/maps/master/india/india-states.geojson'
india_states = gpd.read_file(geojson_url)

# Merge your data with the GeoJSON file
merged_data = india_states.merge(grouped_data, how='left', left_on='state', right_on='end_customer_state_new')

# Create a choropleth map using Plotly Express
fig = px.choropleth(merged_data,
                    geojson=india_states.geometry,
                    locations=merged_data.index,
                    color='quantity',
                    color_continuous_scale='Viridis',
                    labels={'quantity': 'Quantity'},
                    title='Quantity of Sells by State in India',
                    scope='asia',  # Set the map scope to Asia (for India)
                    )

# Customize layout
fig.update_geos(fitbounds='locations', visible=False)  # Fit the map to the specified locations

# Show the plot
fig.show()
```
